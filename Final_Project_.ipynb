```python
import pandas as pd

data = {
    "Order_ID": [101,102,103,104,105,106,107,108,109,110,111,112,113,114,115],
    "Order_Date": ["2024-01-05", "2024-01-08", "2024-01-10", "2024-01-12", "2024-01-15", "2024-01-18", "2024-01-20", "2024-01-22", "2024-01-25", "2024-01-28", "2024-02-02", "2024-02-05", "2024-02-08", "2024-02-10", "2024-02-12"],
    "Customer_Name": ["Ravi", "Priya", "Arun", "Meena", "Karthik", "Divya", "Suresh", "Latha", "Vijay", "Anitha", "Kumar", "Nisha", "Prakash", "Keerthi", "Manoj"],
    "Category": ["Electronics", "Fashion", "Home Decor", "Electronics", "Fashion", "Books", "Electronics", "Home Decor", "Fashion", "Electronics", "Books", "Fashion", "Electronics", "Home Decor", "Electronics"],
    "Product_Name": ["Laptop", "T-Shirt", "Wall Clock", "Smartphone", "Shoes", "Python Guide", "Headphones", "Lamp", "Jeans", "Tablet", "Data Science", "Jacket", "Smart Watch", "Curtains", "Monitor"],
    "Region": ["South", "North", "East", "West", "South", "North", "East", "West", "South", "North", "East", "West", "South", "North", "East"],
    "Quantity": [2, 5, 3, 4, 2, 6, 5, 4, 3, 2, 4, 2, 3, 5, 2],
    "Sales": [120000, 5000, 4500, 80000, 6000, 3000, 10000, 8000, 4500, 30000, 4000, 7000, 15000, 7500, 20000],
    "Profit": [18000, 1200, 900, 15000, 1800, 1000, 2500, 2000, 1100, 6000, 1300, 2200, 3500, 1800, 5000]
}

df = pd.DataFrame(data)

df["Order_Date"]=pd.to_datetime(df["Order_Date"])
df["Month"]=df["Order_Date"].dt.month_name()
monthly_sales=df.groupby("Month")["Sales"].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales Performance')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
category_sales.plot(kind='bar', color='lightcoral')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
profit_by_category.plot(kind='bar', color='lightgreen')
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
